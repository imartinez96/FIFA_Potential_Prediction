# DATA STORYTELLING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa_players = pd.read_csv('data.csv')
```

```python
plt.figure(figsize=(18,10))
sns.countplot(fifa_players['Potential'], palette='rocket')
plt.title('Count Plot of Potential Score of players')
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.heatmap(fifa_players.corr(), annot=True, linewidths = 1);
plt.title('Dataset Heatmap')
plt.show()
```

## Is there any correlation between a players potential and his other skill sets? If yes, which skills tend to have more impact?

We have plotted a heat map above for the entire dataset to find out hidden correlations between different attributes. As can be seen above potential and overall score seem to be correlated with most of the other skill sets. Overall rating is strictly a FIFA game statistic so we cannot use this to predict player potential. 

Similarly, we can also see that age has a negative correlation with skills like sprint speed and acceleration, as expected players at younger age would be more agile. It also has a negative correlation with player potential. This is also expected because as a player ages, there is less room to grow.

It can be stated that other physical attributes of a football player do have an impact on deriving potential score of that player but it is surprising to see how much the mental attributes are so highly correlated to player's potential score and physical statistics. This indicates that soccer performance has a heavy mental side to it.

Lets explore this more.

## Is there any correlation between a players mental attributes and potential score?

```python
#Composure vs Potential
plt.figure(figsize=(20, 15))
sns.regplot(x = 'Composure' , y = 'Potential', data = fifa_players )
plt.title('Composure vs Potential rating')
plt.show()
```

```python
#Reaction Speed vs Potential
plt.figure(figsize=(20, 15))
sns.regplot(x = 'Reactions' , y = 'Potential', data = fifa_players )
plt.title('Reactions vs Potential rating')
plt.show()
```

```python
#Aggression vs Potential
plt.figure(figsize=(20, 15))
sns.regplot(x = 'Aggression' , y = 'Potential', data = fifa_players )
plt.title('Aggression vs Potential rating')
plt.show()
```

The scatter plots demonstrated above does reveal that the potential score of a player is dependent on mentality of the soccer player. It can be figured out that players who can keep good composure throughout the game and are quick to react to game situations, seem to have a high potential score. 

This might be related to their capability to keep emotions in check and being highly focused on the field, that lead to better performance in their physical statistics as well. Thus, it can be concluded that the correlation between a players potential and his mental attribute scores is high.
